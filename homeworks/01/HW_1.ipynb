```python
class HashMap:
    #Мое хранилище ключей и значений для итераторов
        
    class Entry:
        def __init__(self, key, value):
            #TODO store key and value as attributes
            self.key = key
            self.value = value
            
        def get_key(self):
            #TODO return key
            return self.key
            
        def get_value(self):
            #TODO return value
            return self.value
            
        def __eq__(self, other):
            #TODO реализовать функцию сравнения
            #Пояснения: self - текущий объект entry, other - какой-то другой объект
            #Какое условие на равенство объектов в ассоциативном массиве?
            #Если объекты равны, то метод должен возвращать True, иначе False
            if self.key == other.key:
                return True
            else:
                return False
    
    def __init__(self, bucket_num = 64):
        #TODO Создать массив для бакетов
        self.bucket = [[] for __ in range(bucket_num)] #пустой массив заданной длины
        self.bucket_num = bucket_num #будем использовать в хеш функции

    def get(self, key, default_value=None):
        #TODO метод get, возвращающий значение, если оно присутствует, иначе default_value
        init_val = self._get_index(self._get_hash(key))
        for i in range(len(self.bucket[init_val])):
            obj = self.bucket[init_val][i]
            if obj.key == key:
                return obj.value
        return default_value
    
    #мой сборщик значений (ключи/значения) в массив ключей_значений в итераторе
    def up_to_sets(function):
        def funk_arr(self,key,value):
            function(self,key,value)
            for i in range(len(self.KeyValuesIter.array_key_value)):
                if self.KeyValuesIter.array_key_value[i][0] == key:
                    self.KeyValuesIter.array_key_value[i][1] = value
                    return
            key_val = [key,value]
            self.KeyValuesIter.array_key_value.append(key_val)
               
        return funk_arr

    @up_to_sets
    def put(self, key, value):
        #TODO метод put, кладет значение по ключу, в случае, если ключ уже присутствует он его заменяет
        if (type(key) == type(list())) or (type(key) == type(dict())) or (type(key) == type(set())):
            print("invalid type")
            
        init_val = self._get_index(self._get_hash(key))
        x = self.Entry(key,value)
        for i in self.bucket[init_val]:
            if i.__eq__(x):
                i.value = x.value
                return
        self.bucket[init_val].append(x)

    def __len__(self):
        #TODO Возвращает количество Entry в массиве
        sum = 0;
        for i in self.bucket:
            sum += len(i)
        return sum

    def _get_hash(self, key):
        #TODO Вернуть хеш от ключа, по которому он кладется в бакет
        return hash(key)

    def _get_index(self, hash_value):
        #TODO По значению хеша вернуть индекс элемента в массиве
        my_index = hash_value%self.bucket_num
        if my_index < 0:
            my_index = my_index * (-1)
        return my_index
    
    # Итератор для ключей и значений (поставил сюда, чтобы не множить код, сделал сразу универсальным(ключ/значение), по дефолту работает для ключей)
    class KeyValuesIter:
        array_key_value = []
        
        def __iter__(self):
            return self
        
        def __init__ (self,KF = 0):
            self.array = self.array_key_value
            self.limit = len(self.array)
            self.count = 0
            self.KF = KF
        
        def __next__ (self):
            if self.count < self.limit:
                self.count += 1
                return self.array[self.count-1][self.KF]
            else:
                raise StopIteration
        
    def values(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор значений (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.KeyValuesIter(1) # вызов итератора для значений и ключей (с флагом KF = 1, что означает, что в данный момент работаем с значениями)
        
    def keys(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор ключей (на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        return self.KeyValuesIter() # вызов итератора для значений и ключей (с флагом KF = 0, что означает, что в данный момент работаем с ключами)
        
    def items(self):
        #TODO Задание со звездочкой!
        #Должен возвращать генератор или итератор пар ключ и значение (tuples)
        #(на самом деле итератор, но не принципиально, через генераторы проще ИМХО)
        # наследник итератора для значений и ключей
        class ItemsIter(self.KeyValuesIter):
            def __next__(self):
                if self.count < self.limit:
                    self.count += 1
                    ret = (self.array[self.count-1][0], self.array[self.count-1][1])
                    return ret
                else:
                    raise StopIteration
        return ItemsIter()

    def __str__(self):
        #TODO Метод, вызываемый при вызове print, пофантазируйте, что можно написать
        Mystr = "["
        for hash_array in self.bucket:
            for obj in hash_array:
                Mystr += str(obj.get_key()) + ':' + str(obj.get_value()) + ' , '
        Mystr += "]"
        return Mystr
                
```

```python
import random

def test_hashmap_04():
    entries = [(5, 7), ("entries", 56), ("value", 54.), (1000, "t"), (HashMap(10), ()), ({"s":"v"}, {"v":"s"})]
    for k, v in entries:
        entry = HashMap.Entry(k, v)
        assert entry.get_key() == k
        assert entry.get_value() == v
    print("Test 4 part 01 passed")
        
    for i in range(len(entries)):
        entry_one = HashMap.Entry(entries[i][0], entries[i][1])
        for _ in range(10):
            j = random.randint(0, len(entries)-1)
            p = random.randint(0, len(entries)-1)
            entry_two = HashMap.Entry(entries[j][0], entries[p][1])
            if j == i:
                #print(entry_one.__dict__, entry_two.__dict__, i,j,p)
                assert entry_one == entry_two
            else:
                assert entry_one != entry_two
    print("Test 4 part 02 passed")
    
test_hashmap_04()
```
