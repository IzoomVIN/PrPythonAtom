### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crime_decade = crime.set_index('Year')

i=0
n_c_d ={}
while i < len(crime_decade.index):
    
    dec = crime_decade.index[i]
    n_c_d[dec] = {'Population':None}
    ix = crime_decade.index[i]
    
    while (i < len(crime_decade.index)) and (dec % 100 // 10 == crime_decade.index[i] % 100 // 10) and (
        (dec%10 != crime_decade.index[i] % 10) or crime_decade.index[i] % 10 == 0):
        
        if (n_c_d[dec]['Population'] == None) or (n_c_d[dec]['Population'] <  crime_decade.Population[ix]):
            n_c_d[dec]['Population'] = crime_decade.Population[ix]
            
        for j in crime_decade.columns[1:]:
            if j not in n_c_d[dec].keys():
                n_c_d[dec][j] = 0
            n_c_d[dec][j] += crime_decade[j][ix]
        i += 1
```

```python
crime_decade = pd.DataFrame(n_c_d)
crime_decade = crime_decade.T
```

### Step 9. What is the mos dangerous decade to live in the US?

```python
ar = [str(i) for i in crime_decade.columns.values]
crime_decade.sort_values(ar)
#1990
```
