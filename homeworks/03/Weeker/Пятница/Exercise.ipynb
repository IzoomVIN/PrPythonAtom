# Occupation

### Introduction:


### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
datepd = pd.read_csv(url, sep = '|')
date = pd.DataFrame(datepd)
date.head()
```

### Step 3. Assign it to a variable called users.

```python
users = date
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby("occupation").age.mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
us = users[['occupation', 'gender']][users.gender == 'M']
us = us.groupby('occupation').count()
us = us.sort_values('gender', ascending=False)
us.head()
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
min_ = users[['occupation', 'age']].groupby('occupation').min()
max_ = users[['occupation', 'age']].groupby('occupation').max()
inf = min_.join(max_, lsuffix = '_min', rsuffix = '_max')
inf.head()
```

```python
inf = users[['occupation', 'age']]
inf = inf.groupby('occupation').age.agg(['min', 'max'])
inf
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
inf_pt = pd.pivot_table(users, index = ['occupation', 'gender'], values = 'age')
inf_pt.head()
```

### Step 8.  For each occupation present the percentage of women and men

```python
count =  users[['occupation', 'gender']].groupby('occupation').count()

male = users[['occupation', 'gender']][users.gender == 'M']
male = (male.groupby('occupation').count()/count)*100

female = users[['occupation', 'gender']][users.gender == 'F']
female = (female.groupby('occupation').count()/count)*100

info = male.join(female, lsuffix = '_m_percent', rsuffix = '_f_percent')
info.head()
```
